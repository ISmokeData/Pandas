### **What is DataFrame?**
*A Data frame is a two-dimensional data structure, i.e., data is aligned in a tabular fashion in rows and columns. Pandas DataFrame consists of three principal components, the data, rows, and columns.*

```python
import numpy as np
import pandas as pd
```

### **Creating DataFrame**

```python
#  using lists
student_data = [
    [100, 90, 10],
    [90, 70, 7],
    [120, 100, 14],
    [80, 50, 2]
]
pd.DataFrame(student_data, columns=['iq', 'marks', 'package', ])
```

```python
# using dictionary
student_dict = {
    'iq': [100, 90, 80, 120, 0, 0],
    'marks': [80, 70, 100, 90, 0, 0],
    'package': [10, 7, 14, 2, 0, 0]
}

students = pd.DataFrame(student_dict)
students
```

### **Selecting rows from a DataFrame**
- **iloc** - searches using index position
- **loc** - searches using index labels

```python
student_dict = {
    'name': ['nitish', 'rupesh', 'rishabh', 'amit', 'ankita', 'suresh'],
    'iq': [100, 90, 80, 120, 0, 0],
    'marks': [80, 70, 100, 90, 0, 0],
    'package': [10, 7, 14, 2, 0, 0]
}

students = pd.DataFrame(student_dict)
students.set_index('name', inplace=True)
students
```

```python
# single row
movies.iloc[0]
```

```python
# multiple row
movies.iloc[0:5]
```

```python
movies.iloc[0:10:2]
```

```python
students.loc['nitish']
```

```python
students.loc['nitish':'rishabh']
```

```python
# slicing row and columns using iloc
movies.iloc[0:3, 0:3]
```

```python
# slicing row and columns using loc
movies.loc[0:3, 'title_x': 'poster_path']
```

```python
# movies with rating higher than 8 and votes > 10000
movies[(movies['imdb_rating'] > 8) & (movies['imdb_votes'] > 10000)
       ][['title_x', 'imdb_votes', 'imdb_rating']]
```

```python
# Action movies with rating higher than 7.5
mask1 = movies[movies['genres'].str.split('|').apply(lambda x: 'Action' in x)]
mask1[mask1['imdb_rating'] > 7.5]
```

```python
# another method
mask1 = movies['genres'].str.contains('Action')
mask2 = movies['imdb_rating'] > 7.5
movies[mask1 & mask2]
```
