### **Join / Merge**

```python
# inner join
students.merge(regs, how='inner', on='student_id')
```

```python
# outer -> print those value which have common and left side but not right side
course.merge(regs, how='left', on='course_id')
```

```python
# right join
temp_df = pd.DataFrame(
    {
        'student_id': [26, 27, 28],
        'name': ['nitish', 'ankit', 'rahul'],
        'partner': [28, 26, 17]
    }
)
temp_df
```

```python
students = pd.concat([students, temp_df])
students
```

```python
students.merge(regs, how='right', on='student_id')
```

```python
# outer
students.merge(regs, how='outer', on='student_id').tail(10)
```

```python
# 1. find total revenue generated
regs.merge(course, how='inner', on='course_id')['price'].sum()
```

```python
# 2.find month by month revenue
temp_df = pd.concat([nov, dec], keys=['Nov', 'Dec']).reset_index()
temp_df.merge(course, on='course_id').groupby('level_0')['price'].sum()
```

```python
# 3.print the regristration table
# cols - > name -> course -> price
new = regs.merge(students, on='student_id').merge(
    course, on='course_id')[['name', 'course_name', 'price']]
new
```

```python
# 4. plot bar chart for revenue/course
regs.merge(course, on='course_id').groupby(
    'course_name')['price'].sum().plot(kind='bar')
```

```python
common_student_id = np.intersect1d(nov['student_id'], dec['student_id'])
common_student_id
```

```python
students[students['student_id'].isin(common_student_id)]
```

```python
# 6. find course that got not enrollment
course_id_list = np.setdiff1d(course['course_id'], regs['course_id'])
course[course['course_id'].isin(course_id_list)]
```

```python
# 7. find student who did not enroll into any courses
student_id_list = np.setdiff1d(students['student_id'], regs['student_id'])
students[students['student_id'].isin(student_id_list)]
```

```python
# 8. print students name -> partner name for all enrolled students
# self.join
students.merge(students, how='inner', left_on='partner',
               right_on='student_id')[['name_x', 'name_y']]
```

```python
# 9. find top 3 students who did most number enrollments
regs.merge(students, on='student_id').groupby(
    ['student_id', 'name'])['name'].count().sort_values(ascending=False).head(3)
```

```python
# 10. money on course
regs.merge(students, on='student_id').merge(
    course, on='course_id').groupby(['student_id', 'name'])['price'].sum().sort_values(ascending=False).head(3)
```

```python
# alternate syntax for merge
pd.merge(students, regs, how='inner', on='student_id')
```
