### **Series is 1D and DataFrame are 2D objects**
- but why?
- and what exactly is index?

```python
# can we have multiple index? let's try
index_val = [('cse', 2019), ('cse', 2020), ('cse', 2021), ('cse', 2022),
             ('ece', 2019), ('ece', 2020), ('ece', 2021), ('ece', 2022)]
a = pd.Series([1, 2, 3, 4, 5, 6, 7, 8], index=index_val)
a
```

```python
a[('cse', 2022)]
```

```python
# The solution -> multindex series(also known as hierarchical indexing)
# multiple index within a single index
```

```python
# how to create multindex object
# 1. pd.Multindex.from_tuples()
# 2. pd.Multindex.from_product()
```

```python
index_val = [('cse', 2019), ('cse', 2020), ('cse', 2021), ('cse', 2022),
             ('ece', 2019), ('ece', 2020), ('ece', 2021), ('ece', 2022)]
multindex = pd.MultiIndex.from_tuples(index_val)
multindex
```

```python
multindex.levels[0]
```

```python
multindex.levels[1]
```

```python
# another method
pd.MultiIndex.from_product((['cse', 'ece'], [2019, 2020, 2021, 2022]))
```

```python
# creating series with multindex object
s = pd.Series([1, 2, 3, 4, 5, 6, 7, 8], index=multindex)
s
```

```python
s[('cse', 2022)]
```

```python
s['cse']
```

```python
# unstack
temp = s.unstack()
temp
```

```python
temp.stack()
```

### **Long Vs Wide Data**

Wide format is where we have a single now for every data point with multiple columns to hold the values of various attributes.

Long format is where for each data point we have as many rows as the number of attributes and each row contains the value of a particular attributes for a given data point.

```python
# melt -> wide to long
pd.DataFrame({'cse': [120]}).melt()
```

```python
pd.DataFrame({'cse': [120], 'ece': [100], 'mech': [50]}).melt(
    var_name='branch', value_name='num_student')
```

```python
pd.DataFrame(
    {
        'branch': ['cse', 'ece', 'mech'],
        '2020': [100, 150, 60],
        '2021': [120, 130, 80],
        '2022': [150, 140, 70]
    }
).melt()
# its not working well
```

```python
pd.DataFrame(
    {
        'branch': ['cse', 'ece', 'mech'],
        '2020': [100, 150, 60],
        '2021': [120, 130, 80],
        '2022': [150, 140, 70]
    }
).melt(id_vars=['branch'], var_name='year', value_name='student')
```

```python
death = pd.read_csv('time_series_covid19_deaths_global.csv')
confirm = pd.read_csv('time_series_covid19_confirmed_global.csv')
```

```python
death.head()
```

```python
death_new = death.melt(id_vars=['Province/State', 'Country/Region',
                                'Lat', 'Long'], var_name='date', value_name='death')
death_new.head()
```

```python
confirm_new = death.melt(id_vars=['Province/State', 'Country/Region',
                                  'Lat', 'Long'], var_name='date', value_name='confirm')
confirm_new.head()
```

```python
confirm_new.merge(death_new, on=['Province/State', 'Country/Region', 'Lat',
                                 'Long', 'date'])[['Country/Region', 'date', 'confirm', 'death']]
```
